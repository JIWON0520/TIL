#### 결측치 처리

```python
#,결측치 처리를 위해 학습 데이터 세트와 테스트 데이터 세트를 합침
all_data=pd.concat([train,test],ignore_index=True)
all_data.shape
```

```python
#count 피처 삭제
all_data=all_data.drop('count',axis=1)

all_data.isnull().sum()
```

```python
#결측치 개수가 5개 이상인 로우 삭제
#all_data.dropna(thresh=5,axis=0,inplace=True)
#all_data.isnull().sum()
```

```python
#피처들의 평균값
all_data.mean()
```

```python
#groupby로 시간으로 그룹을 지은 뒤 피처들의 평균값
#groupby하지 않은 평균과는 많이 차이가 있음
all_data.groupby('hour').mean()
```

```python
all_data[all_data['hour']==1][['hour_bef_ozone','hour_bef_pm10']].tail(5)
```

```python
import numpy as np

#hour이 1일때 hour_bef_ozone,hour_bef_pm10,hour_bef_pm2.5은 모두 null값임
#그래서 앞 뒤 시간의 평균값으로 대체
all_data[all_data['hour']==1]=all_data[all_data['hour']==1].fillna({'hour_bef_ozone':np.mean([0.032472,0.030132]),
                                'hour_bef_pm10':np.mean([48.786517,50.747253]),
                                'hour_bef_pm2.5':np.mean([28.329545,28.777778])})
all_data.isnull().sum()
```

```python
#나머지 결측치들은 각 시간대의 평균 값으로 대체
all_data=all_data.groupby('hour').apply(lambda x:x.fillna(x.mean()))
```

```python
all_data.isnull().sum()
```

```python
all_data.head(5)
```

```python
index=[index[1] for index in all_data.index.tolist()]
index
```

```python
all_data.index=index
all_data=all_data.sort_index()
all_data.head(5)
```

```python
#다시 학습데이터 세트와 테스트 데이터 세트를 분리
train=all_data.iloc[:1459,:].merge(train[['id','count']],on='id',how='left')
test=all_data.iloc[1459:,:]
train.shape, test.shape
```

```python
train.head(5)
```

```python
train.isnull().sum()
```

#### DecisionTree를 이용한 회귀 예측

```python
from sklearn.tree import DecisionTreeRegressor

X_train=train.drop('count',axis=1)
Y_train=train['count']

dt=DecisionTreeRegressor()
dt.fit(X_train,Y_train)
```

```python
pred=dt.predict(test)
```

#### submission 파일 생성

```python
submission=pd.read_csv("./따릉이/submission.csv")
submission['count']=pred
submission.head()
```

```python
submission.to_csv('./submission1.csv',index=False)
```
